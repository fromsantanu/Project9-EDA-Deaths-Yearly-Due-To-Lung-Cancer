# Incidents of Deaths due to Lung Cancer

Deaths due to Lung Cancer by Risk Factors of Smoking, Aloholism and being a COPD patient (2018-2023)

### Data Preparation (Synthetic)

```python
import pandas as pd

# Data for deaths due to lung cancer over the last six years
data_lung_cancer = {
    "Year": ["2018", "2019", "2020", "2021", "2022", "2023"],
    "Smokers": [13000, 13800, 15500, 15500, 16200, 17000],
    "Alcoholics": [8020, 8400, 9000, 9200, 10000, 10200],
    "COPD": [5000, 5600, 5700, 5800, 5900, 6000]
}

# Creating a DataFrame
df_lung_cancer = pd.DataFrame(data_lung_cancer)

# Saving the DataFrame to a CSV file
df_lung_cancer.to_csv('lung_cancer_deaths.csv', index=False)

print("Data saved to lung_cancer_deaths.csv")
```

### Plotting the chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading the data from the CSV file
df_lung_cancer = pd.read_csv('lung_cancer_deaths.csv')

# Plotting the bar chart
plt.figure(figsize=(12, 7))
bar_width = 0.2

# Positions of the bars on the x-axis
r1 = range(len(df_lung_cancer))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Making the bar plot
plt.bar(r1, df_lung_cancer["Smokers"], color='Red', width=bar_width, edgecolor='grey', label='Smokers')
plt.bar(r2, df_lung_cancer["Alcoholics"], color='Yellow', width=bar_width, edgecolor='grey', label='Alcoholics')
plt.bar(r3, df_lung_cancer["COPD"], color='Grey', width=bar_width, edgecolor='grey', label='COPD Patients')

# Adding xticks
plt.xlabel('Year', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(df_lung_cancer))], df_lung_cancer["Year"])

# Adding labels and title
plt.ylabel('Number of Deaths', fontweight='bold')
plt.title('Deaths due to Lung Cancer by Risk Factor (2018-2023)', fontweight='bold')

# Adding legend
plt.legend()

# Display the chart
plt.tight_layout()
plt.show()
```
